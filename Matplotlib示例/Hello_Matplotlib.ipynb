# 1. Matplotlib 是什么
- Matplotlib一个专门用于开发2D(3D)图表的工具，以渐进和交互方式实现数据可视化.
- JS脚本D3、echart等都可以实现表现丰富的数据可视化效果，但根据奥卡姆剃刀原理（如无必要、勿增实体）的原则，Matplotlib使用更方便，以最基础、原始的方式呈现数据可视化效果。

# 2. Matplotlib的三层结构
- 1.容器层
    - 画板层（canvas）
    - 画布层（Figure): plt.figure()
    - 绘图区/坐标系(Axes): plt.subplots()
        - 2.辅助现实层: 主要包挎哦Axes外观（facecolor）、边框线（spines）、坐标轴（axis）、坐标轴名称（axis lable）、坐标轴刻度（tick）、作弊啊轴刻度标签（tick label）、网格线（grid）、图例（legend）、标题（title）等内容
        - 3.图像层: Axes内通过折线图（plot）、散点图（scatter）、柱状图（bar）、直方图（histogram）、饼图（pie）等函数绘出图像

# 3. Matplotlib绘图
### 3.1 折线图（plot）

* 图形格式风格

| 颜色字符   |   线风格字符     |   标记风格字符     |
| :-------- | -------:      |     -------:      |
| r（红色）  |  - （实线）     |     . （点）      |
| g（绿色）  |  --（虚线）     |     ,（像素点）    |
| b（蓝色）  |  -.（点划线）   |     o（圆点）      |
| w（白色）  |  : （点虚线）   |     v（下三角点）   |
| c（青色）  |      ...      |     ^（上三角点）   |
| m（洋红）  |      ...      |     <（左三角点）   |
| y（红色）  |      ...      |     >（右三角点）   |
| k（黑色）  |      ...      |     1（下三叉点）   |
| ...       |      ...      |     2（上三叉点）   |
| ...       |      ...      |     3（左三叉点）   |
| ...       |      ...      |     4（右三叉点）   |
| ...       |      ...      |     s（正方点）   |
| ...       |      ...      |     p（五角点）   |
| ...       |      ...      |     *（星形点）   |
| ...       |      ...      |     h（六边形点1）|
| ...       |      ...      |     H（六边形点2）|
| ...       |      ...      |     +（加号点）|
| ...       |      ...      |     x（乘号点）|
| ...       |      ...      |     D（实心菱形点）|
| ...       |      ...      |     d（瘦菱形点）|
| ...       |      ...      |     _（横线点）|

```python
import matplotlib.pyplot as plt
import random
x = range(24)
y_shanghai = [random.uniform(15,24) for i in x]
y_beijing = [random.uniform(0,10) for i in x]
# 创建画布 plt.figure(figsize(), dpi =)
plt.figure(figsize = (6, 3), dpi = 96)
# 绘制图像
plt.plot(x, y_shanghai, label="上海")
plt.plot(x, y_beijing, color="r", linestyle="--", label="北京", marker=".")
plt.legend(['上海','北京'])
# 显示x, y轴刻度
x_label = ["{}时".format(i) for i in x]
plt.xticks(x[::4], x_label[::4])
plt.yticks(range(0,40,5))
# 显示网格
plt.grid(True, linestyle="--", alpha=0.5)
# 添加描述信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("上海与北京一天24小时温度变化状况")
# 显示图像,这一步需要注意，会释放figture资源
plt.show()
```

### 3.2 创建多个绘图区绘制其他类型图（plt.subplots）

* matplotlib.pyplot.subplots(nrows=1, ncols=1, **fig_kw) 创建一个带有多个axes的图

    #### 3.2.1 散点图 (scatter)

    #### 3.2.2 柱状图 (bar)

    #### 3.2.3 直方图（hist）

    #### 3.2.4 饼图（pie）


```python
import matplotlib.pyplot as plt
import random
# 创建画布 plt.figure(figsize(), dpi =)
fig=plt.figure(figsize = (12, 12), dpi = 96)
axes=fig.subplots(nrows=2, ncols =2)


# 绘制散点图
r0 = range(30)
x0 = [random.uniform(0,20) for i in r0]
y0 = [random.uniform(0,20) for i in r0]
x1 = [random.uniform(0,20) for i in r0]
y1 = [random.uniform(0,20) for i in r0]

 
axes[0,0].scatter(x1,y1)
axes[0,0].legend(['上海','北京'])

# 显示axes[0,0] x, y轴刻度
axes[0,0].set_xticks(range(0,25,5))
axes[0,0].set_yticks(range(0,25,5))
x_labels = ["{}厘米".format(i) for i in r0]
y_labels = ["{}厘米".format(i) for i in r0]
axes[0,0].set_xticklabels(x_labels[::5])
axes[0,0].set_yticklabels(y_labels[::5])
# 显示网格
axes[0,0].grid(True, linestyle="--", alpha=0.5)
# 添加描述信息
axes[0,0].set_xlabel("x坐标")
axes[0,0].set_ylabel("y坐标")
axes[0,0].set_title("散点图")


# 绘制柱状图
x2 = ["中国","美国","德国","日本"]
y2 = [6500,3200,1200,800]
z2 = [1200,2400,500,300]
# 显示axes[0,1] x, y轴刻度
axes[0,1].bar(range(len(x2)),y2,width=0.25)
axes[0,1].bar(range(len(x2)),z2,width=0.25)
axes[0,1].set_xticks(range(len(x2)))
axes[0,1].set_xticklabels(x2)
axes[0,1].legend(['综合实力','军事实力'])
# 添加描述信息
axes[0,1].set_xlabel("国家")
axes[0,1].set_ylabel("综合实力与军事实力")
axes[0,1].set_title("柱状图")

# 绘制直方图
x3 = [117,138,112,124,129,141,128,122,112,90,95,70,100,125,107,161,170,117,138,112,124,129,141,128,122,112,90,95,80,117,149,99,103]
group_num = (max(x3)-min(x3))//10
axes[1,0].hist(x3,bins = group_num,density = False)
axes[1,0].set_xticks(range(min(x3),max(x3),10))
# 显示网格
axes[1,0].grid(True, linestyle="--", alpha=0.5)
# 添加描述信息
axes[1,0].set_xlabel("组距")
axes[1,0].set_ylabel("频数")
axes[1,0].set_title("直方图")

# 绘制饼图
x4 = ["腾讯","网易","完美世界","搜狐畅游","西山居","其他"]
y4 = [5400,3200,1670,852,1105,2460]

axes[1,1].pie(y4, labels=x4, autopct="%1.2f%%")
axes[1,1].axis("equal")
axes[1,1].legend(loc='upper left')
# 添加描述信息
axes[1,1].set_title("市场份额饼图")
# 显示图像,这一步需要注意，会释放figture资源
plt.show()
```
