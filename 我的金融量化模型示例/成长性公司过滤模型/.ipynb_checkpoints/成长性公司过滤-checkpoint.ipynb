# 成长性公司过滤
### 三个标准：
* 最近一年公司主营业务收入同比增长30%以上（年报）
* 最近一年季度主营收入环比增长30%以上（滚动）
* 三年公司主营收入复合增长率30%以上

### 附加条件
* 三年归母扣非净利润率增长0%以上

```python
#token : 50897e472b76080caf1a0153f282761bde5f99a09280593344002a96
import tushare as ts
import pandas as pd
import numpy as np
import datetime
import time
from scipy.stats.mstats import gmean

api_counter = 0

check_start_data = '20190630'
check_end_date = '20200930'
years_array = ['20191231', '20181231', '20171231']

def check_condition( stock_code ):
    df_condition = pro.query('fina_indicator', ts_code = stock_code, start_date = check_start_data, end_date = check_end_date, fields = 'ts_code, ann_date, end_date, or_yoy, q_sales_qoq')
    df_condition.replace(to_replace=[None], value=np.nan, inplace=True) 
    q_sales_qoq = gmean(df_condition['q_sales_qoq'].values/100.0 + 1.0)
    q_sales_qoq = (q_sales_qoq-1)*100

    or_yoy_grow_array = []
    dt_netprofit_grow_array = []
    for year in years_array:
        df_condition3 = pro.query('fina_indicator', ts_code = stock_code, period = year, fields = 'ts_code, ann_date, end_date, or_yoy, dt_netprofit_yoy')
        df_condition3.replace(to_replace=[None], value=np.nan, inplace=True)
        if(df_condition3.index.size >0):
            or_yoy_year_grow_rate = df_condition3.loc[0, 'or_yoy']
            or_yoy_grow_array.append(1.0 + or_yoy_year_grow_rate/100.0)
            dt_netprofit_grow_rate = df_condition3.loc[0, 'dt_netprofit_yoy']
            dt_netprofit_grow_array.append(1.0 + dt_netprofit_grow_rate/100.0)
    or_yoy = or_yoy_grow_array[0]
    or_yoy = (or_yoy-1)*100
    or_yoy_year_grow_rate = gmean(or_yoy_grow_array)
    or_yoy_year_grow_rate = (or_yoy_year_grow_rate-1)*100
    dt_netprofit_grow_rate = gmean(dt_netprofit_grow_array)
    dt_netprofit_grow_rate = (dt_netprofit_grow_rate-1)*100
    
    print("------")
    print(or_yoy)
    print(q_sales_qoq)
    print(or_yoy_year_grow_rate)
    print(dt_netprofit_grow_rate)
    print("------")
    if(or_yoy >= 30 and q_sales_qoq >= 10 and or_yoy_year_grow_rate >=30 and dt_netprofit_grow_rate > 0):
        return stock_code
    else:
        return -1

#初始化tushare的token
pro = ts.pro_api('50897e472b76080caf1a0153f282761bde5f99a09280593344002a96')

#查询当前所有正常上市交易的股票列表
stockdata = pro.query('stock_basic', exchange='', list_status='L', fields='ts_code,symbol,name')
# 从dataframe中取出ts_code
ts_code = stockdata['ts_code']
stocks = []

i = 0
progress = 0
for stock_code in ts_code:
    i+=1
    ret = check_condition(stock_code)
    print(stock_code)
    if(ret != -1):
        stocks.append(ret)
    # 打印进度
    # 一共3922支股票
    progress = int((i+1)/3.922)
    print('\r 当前数据提取进度：{0}{1}%'.format('▉'*progress,(progress*0.1)), end='')
    time.sleep(3)
print(' Tushare数据提取完毕')

#筛选后的股票
final_stocks = stockdata.loc[stockdata['ts_code'].isin(stocks)]
print(final_stocks)

final_stocks.to_excel("output/20200810.xlsx")
print("导出到output/20200810.xlsx，成功！！")
```
